```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
from IPython.display import display 

# Load csv files
schools_complete = Path("Resources/schools_complete.csv")
students_complete = Path("Resources/students_complete.csv")

# Read csv files to DataFrames
school_df = pd.read_csv(schools_complete)
student_df = pd.read_csv(students_complete)

#Display DataFrames
display(school_df)
display(student_df)
```

```python
# Calculate total schools, total students, total budget, average reading score, average math score, % passing reading, % passing math, % passing both
total_schools = len(school_df["school_name"])

total_students = school_df["size"].sum()

total_budget = school_df["budget"].sum()

average_reading_score = student_df["reading_score"].mean()

average_math_score = student_df["math_score"].mean()

passing_reading_percentage = (student_df["reading_score"] >= 70).sum()/total_students * 100

passing_math_percentage =(student_df["math_score"] >= 70).sum()/total_students * 100

passing_both_percentage = ((student_df["reading_score"] >=70) & (student_df["math_score"] >= 70)).sum()/total_students * 100
```

```python
# Create a dictionary from the district statistics, then display in a DataFrame
district_statistic = {
    "Total Schools": total_students,
    "Total Budget": total_budget,
    "Average Reading Score": average_reading_score,
    "Average Math Score": average_math_score,
    "% Passing Reading": passing_reading_percentage,
    "% Passing Math": passing_math_percentage,
    "% Passing Both": passing_both_percentage
}

district_statistic_df = pd.DataFrame([district_statistic])
district_statistic_df
```

```python
# Find each school's School name, School type, Total students, Total school budget, Per student budget, Average math score, Average reading score, % passing math, % passing reading
school_type = school_df.set_index(["school_name"])["type"]

per_school_count = school_df["size"]

per_school_budget = school_df["budget"]

per_school_capita = per_school_budget/per_school_count

per_school_math = student_df.groupby(["school_name"])["math_score"].mean()

per_school_reading = student_df.groupby(["school_name"])["reading_score"].mean()

school_students_passing_reading =(student_df["reading_score"] >= 70).sum()/per_school_count * 100

school_students_passing_math = (student_df["math_score"] >= 70).sum()/per_school_count * 100
```

```python
# Create a Dataframe to contain the above school information
school_summary = {
    "School Type": school_type,
    "Total Students": per_school_count,
    "Total School Budget": per_school_budget,
    "Average Reading Score": per_school_reading,
    "Average Math Score": per_school_math,
    "% Passing Reading": school_students_passing_reading,
    "% Passing Math": school_students_passing_math,
    "% Passing Both": school_students_passing_both
}

school_summary_df = pd.DataFrame([school_summary])
school_summary_df
```
